```python
# Descobrindo o menor valor de w_ nos arquivos .txt
```

```python
import os

# Caminho da pasta que contém somente arquivos .txt
folder_path = '../../midis/txt'

# Lista para armazenar os valores extraídos de w_XXX
w_values = []

for filename in os.listdir(folder_path):
    # Considera somente arquivos que terminam com .txt
    if filename.endswith('.txt'):
        file_path = os.path.join(folder_path, filename)

        with open(file_path, 'r', encoding='utf-8') as f:
            lines = f.read().splitlines()
            for line in lines:
                # Quebra a linha em tokens
                tokens = line.strip().split()
                if not tokens:
                    # Linha vazia, pula
                    continue
                # Pega o último token
                last_token = tokens[-1]

                # Verifica se ele começa com w_
                if last_token.startswith('w_'):
                    try:
                        # Separa em ["w", valor] e converte o valor para inteiro
                        w_value = int(last_token.split('_')[1])
                        w_values.append(w_value)
                    except ValueError:
                        # Se não for possível converter, apenas ignora
                        pass

# Calcula o menor valor de w_ encontrado
if w_values:
    threshold = min(w_values)
    print(f"O menor valor encontrado de w_ é: {threshold}")
    print("Use esse valor como seu threshold para remover tokens w_ acima disso.")
else:
    print("Não foram encontrados tokens w_ nos arquivos desta pasta.")
```

```python
import os

# Pasta de origem (onde estão os arquivos .txt)
input_folder = '../../midis/txt'

# Pasta de destino (onde vamos salvar os novos arquivos, sem remover os originais)
output_folder = '../../midis/w_txt'

# Garante que a pasta de destino exista
os.makedirs(output_folder, exist_ok=True)

for filename in os.listdir(input_folder):
    if filename.endswith('.txt'):
        input_file_path = os.path.join(input_folder, filename)
        output_file_path = os.path.join(output_folder, filename)

        with open(input_file_path, 'r', encoding='utf-8') as f_in:
            lines = f_in.read().splitlines()

        new_lines = []
        for i, line in enumerate(lines):
            tokens = line.strip().split()
            if not tokens:
                # Linha vazia
                new_line = ""
            else:
                # Se o último token começar com "w_", removemos
                if tokens[-1].startswith('w_'):
                    tokens = tokens[:-1]
                new_line = " ".join(tokens)

            # Se não for a primeira linha (i > 0), adiciona um espaço no começo
            if i > 0:
                new_line = " " + new_line

            new_lines.append(new_line)

        # Agora gravamos o resultado num NOVO arquivo, sem sobrescrever o original
        with open(output_file_path, 'w', encoding='utf-8') as f_out:
            for new_line in new_lines:
                f_out.write(new_line + "\n")

print("Finalizado! Novos arquivos sem o último token 'w_' foram criados em:", output_folder)
```

```python
import os

# Pasta de origem com arquivos .txt
input_folder = '../../midis/w_txt'

# Pasta de destino (onde vamos salvar os novos arquivos .txt)
output_folder = '../../midis/q_txt'
os.makedirs(output_folder, exist_ok=True)

for filename in os.listdir(input_folder):
    if filename.endswith('.txt'):
        input_file_path = os.path.join(input_folder, filename)
        output_file_path = os.path.join(output_folder, filename)

        # Quadrante de emoção vem dos 2 primeiros caracteres do arquivo (ajuste conforme desejar)
        quadrant = filename[:2]  # Ex.: "Q1", "Q2", etc.

        # Lê todas as linhas do arquivo
        with open(input_file_path, 'r', encoding='utf-8') as f_in:
            lines = f_in.read().splitlines()

        new_lines = []
        for line in lines:
            tokens = line.split()

            if tokens:
                # Insere o quadrante como 2º token (se houver pelo menos 1 token)
                tokens.insert(1, quadrant)

                # Insere o quadrante a cada 256 tokens dentro dessa linha
                i = 256
                while i < len(tokens):
                    tokens.insert(i, quadrant)
                    i += 257

            # Reconstrói a linha
            new_line = " ".join(tokens)
            new_lines.append(new_line)

        # Agora gravamos cada linha processada num novo arquivo
        with open(output_file_path, 'w', encoding='utf-8') as f_out:
            for line in new_lines:
                f_out.write(line + "\n")

        print(f"Processado: {filename} -> {output_file_path}")

print("Concluído! Arquivos com quadrante inserido a cada 256 tokens (e como 2º token) foram gerados em:", output_folder)
```
